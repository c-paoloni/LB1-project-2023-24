# **Performance Evalutation of the HMM profile created from Multiple Structural Alignment**

```python
import pandas as pd
import numpy as np
from sklearn.metrics import confusion_matrix, classification_report, matthews_corrcoef,roc_auc_score
import seaborn as sns
import matplotlib.pyplot as plt
```

#### Function used to obtain predictions

```python
def get_preds(x,th):
    pred=[]
    for value in x:
        if value <= th:
            pred.append(1)
        else:
            pred.append(0)
    predicted = np.array(pred)
    return predicted
```

# 1. Optimization 

## 1.1 Subset1 as optimization set & subset2 as validation set

#### Creation of a DataFrame from Subset1

```python
opt1_df= pd.read_csv("opt1_str.txt", sep= '\t' , header= None, names=['ID','E_value','Label'])
```

```python
opt1_df.head()
```

#### Creation of a DataFrame from Subset2

```python
val1_df= pd.read_csv("opt2_str.txt", sep= '\t' , header= None, names=['ID','E_value','Label'])
```

```python
val1_df.head()
```

#### Creation of two optimization arrays containing the E-values and the actual labels respectively

```python
X_opt1= opt1_df["E_value"].to_numpy()
Y_opt1= opt1_df["Label"].to_numpy()
```

```python
X_opt1
```

```python
Y_opt1
```

#### Creation of two validation arrays containing the E-values and the actual labels respectively

```python
X_val1= val1_df["E_value"].to_numpy()
Y_val1= val1_df["Label"].to_numpy()
```

```python
X_val1
```

```python
Y_val1
```

#### Obtaining predictions and classification report using various E-value tresholds

```python
for th in [0.001, 0.0001, 0.00001, 0.000001, 0.0000001, 0.00000001, 0.000000001, 0.0000000001, 0.00000000001, 0.000000000001, 0.0000000000001, 0.00000000000001, 0.000000000000001]:
    Y_pred1= get_preds(X_opt1, th)
    cm = confusion_matrix (Y_opt1, Y_pred1)
    TN, FP, FN, TP= cm.ravel()
    plt.figure(figsize=(6, 4))
    sns.heatmap(cm, annot=True, cmap='Blues', fmt='g', cbar=False, annot_kws={"size": 14}, xticklabels=['Classe 0', 'Classe 1'], yticklabels=['Classe 0', 'Classe 1'])
    plt.xlabel('Predicted')
    plt.ylabel('Actual')
    titolo= 'Confusion Matrix with th='+ str(th)
    plt.title(titolo+'\nTP={}, TN={}, FP={}, FN={}'.format(TP,TN,FP,FN))
    plt.show()
    print(classification_report (Y_opt1, Y_pred1))
    print("MCC: ",  matthews_corrcoef(Y_opt1, Y_pred1))
            
```

#### Validation of the best threshold

```python
th = 0.00000001

Y_pred1= get_preds(X_val1, th)
cm = confusion_matrix (Y_val1, Y_pred1)
TN, FP, FN, TP= cm.ravel()
plt.figure(figsize=(6, 4))
sns.heatmap(cm, annot=True, cmap='Blues', fmt='g', cbar=False, annot_kws={"size": 14}, xticklabels=['Classe 0', 'Classe 1'], yticklabels=['Classe 0', 'Classe 1'])
plt.xlabel('Predicted')
plt.ylabel('Actual')
titolo= 'Confusion Matrix with th='+ str(th)
plt.title(titolo+'\nTP={}, TN={}, FP={}, FN={}'.format(TP,TN,FP,FN))
plt.show()
print(classification_report (Y_val1, Y_pred1))
print("MCC: ",  matthews_corrcoef(Y_val1, Y_pred1))
```

#### Computation of the ROC curve


```python
TPR_list = []
FPR_list = []

# populate TPR_list and FPR_list
for th in [0.001, 0.0001, 0.00001, 0.000001, 0.0000001, 0.00000001, 0.000000001, 0.0000000001, 0.00000000001, 0.000000000001, 0.0000000000001, 0.00000000000001, 0.000000000000001]:
    Y_pred1 = get_preds(X_opt1, th)
    cm = confusion_matrix(Y_opt1, Y_pred1)
    TN, FP, FN, TP = cm.ravel()
    
    TPR_current = TP / (TP + FN)
    FPR_current = FP / (FP + TN)
    
    TPR_list.append(TPR_current)
    FPR_list.append(FPR_current)

# Plot della curva ROC
plt.plot(FPR_list, TPR_list, linestyle='-', color='r', linewidth=3)

# Linea diagonale (classificatore casuale)
plt.plot([0, 1], [0, 1], linestyle='--', color='cyan')

# Calcola il massimo TPR e il suo indice
max_TPR = max(TPR_list)
max_TPR_index = TPR_list.index(max_TPR)

# Plot the horizontal line at the maximum TPR achieved
plt.axhline(y=max_TPR, color='b', linestyle='--', label=f'Maximum TPR: {max_TPR:.2f}')

# Etichette e titolo
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('ROC Curve')

# Limiti degli assi x e y
plt.xlim([0, 1])
plt.ylim([0, 1.05])
plt.grid(True)
plt.legend()
plt.show()
```
